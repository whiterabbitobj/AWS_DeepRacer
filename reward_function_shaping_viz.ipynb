```python
import numpy as np
import matplotlib.pyplot as plt
import importlib
import math
import numpy as np
```

```python
e = 2.71828
pi = 3.14159
sqrt_2pi = 2.5066
std_dev = .25
mean = 0

first_term = 1/(std_dev*sqrt_2pi)
exp_denom = 2*(std_dev**2)
```

```python
# POSITION SMOOTHNESS REWARD CURVE
x = np.linspace(-1,1,100)
exp_denom = 2*(std_dev**1.5)
plt.plot(x, first_term*(e**-(((abs(x)-mean)**3)/exp_denom))/1.6)
```

```python
# STEERING SMOOTHNESS REWARD CURVE
x = np.linspace(-1,1,100)
exp_denom = 2*(std_dev**2.2)
plt.plot(x, first_term*(e**-(((abs(x)-mean)**3)/exp_denom))/1.6)
```

```python
# SPEED SMOOTHNESS REWARD CURVE
x = np.linspace(0,1,100)
exp_denom = 2*(std_dev**1)
multiplier = 1.6
plt.plot(x, multiplier * (-first_term*(e**-(((abs(x)-mean)**3)/exp_denom))/1.6 + 1))
```

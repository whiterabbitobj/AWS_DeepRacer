```python
ylow = 0
yhigh = .3

xlow = 0
xhigh = 1.1

box = [xlow, xhigh, ylow, yhigh]

def e(x, box):   
    ylow = box[2]
    yhigh = box[3]
    
    xlow = box[0]
    xhigh = box[1]
    
    steepmult = 8
    
    steepness = steepmult / (xhigh - xlow)
    offset = (xhigh + xlow) / 2
    midpoint = yhigh - ylow
    
    x = steepness * (x - offset)
    curve = ylow + midpoint / (1 + np.exp(x))
    return curve

x = np.linspace(-10, 10, 150)
plt.figure(1)
plt.subplot(211)
plt.plot(x, e(x, box))
x = np.linspace(xlow-.5, xhigh+.5, 50)
plt.subplot(212)
curve = e(x, box)
plt.yticks(np.linspace(curve.min(), curve.max(), 8))
plt.plot(x, curve)
```
